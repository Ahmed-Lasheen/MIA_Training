FILTERING

```python
import pandas as pd
import numpy as np
import os
df = pd.read_csv(r"E:\Ahmed\MIA\coding\PPython\task4.1\Weather Dataset.csv")


df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])   
df = df.sort_values(by=['Formatted Date'], ascending=True)
df = df[df['Visibility (km)']>=0]
df = df[df['Humidity']>=0]
df = df[df['Pressure (millibars)']>900]  
df = df[df['Temperature (C)']>-50] 
df = df[df['Loud Cover']==0]  
df = df[df['Wind Bearing (degrees)']<=360]      

df.to_csv('filtered_Weather Dataset.csv',index=False)
```

PLOTTINNG

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"E:\Ahmed\MIA\coding\PPython\task4.1\filtered_Weather Dataset.csv")

Temprature = df['Temperature (C)']
Date = df['Formatted Date']

plt.plot(Date, Temprature, color='red')
plt.xlabel('Date')
plt.ylabel('Temprature')
plt.title('Weather')
plt.show()
```
